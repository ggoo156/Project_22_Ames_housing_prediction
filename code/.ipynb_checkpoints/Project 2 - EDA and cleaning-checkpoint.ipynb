### Datatype of each column

```python
df.dtypes
```

### Number of null values (percentage) in each column

```python
# Make a dataframe of null values.
null_values = pd.DataFrame(df.isnull().sum(), columns=['number_of_null_values'])

# Add null value percentage column.
null_values['null_val_percentage'] = null_values.apply(lambda row: row['number_of_null_values'] / 2051 * 100, axis = 1)

# Append datatype column.
null_values['datatype'] = pd.DataFrame(df.dtypes)

null_values
```

### Filter the columns with null value.

```python
null_values[ null_values['number_of_null_values'] != 0 ]
null_values[ null_values['number_of_null_values'] != 0 ].sort_values(by='null_val_percentage', ascending = False)
```

#### Pool QC, Alley, and Misc Feature have too many null values which nulll percentage is over 90%. We will exclude these features since it won't impact the result  significantly.

```python
df.drop(columns=['Pool QC', 'Alley', 'Misc Feature'], inplace = True)
df.head()
```

### Build a function that fill null value.

```python
def fill_null_values(df, column_name, value):
    df[column_name] = df[column_name].fillna(value)
    return df
```

### Fill right values instead of null values by datatype.

```python
# Fill 0 for the columns with float datatype.
fill_null_values(df, ['Lot Frontage', 
                     'Bsmt Full Bath', 
                     'Bsmt Half Bath', 
                     'Garage Cars', 
                     'Garage Area', 
                     'Mas Vnr Area', 
                     'BsmtFin SF 1',
                     'BsmtFin SF 2',
                     'Bsmt Unf SF',
                     'Total Bsmt SF',
                     'Garage Yr Blt'], 0)

# Fill 'None' according to the data description.
fill_null_values(df, 'Mas Vnr Type', 'None')

# Fill 'NA' for the columns with object datatype.
fill_null_values(df, ['Bsmt Qual', 
                     'Bsmt Cond', 
                     'Bsmt Exposure',
                     'BsmtFin Type 1', 
                     'BsmtFin Type 2',
                     'Fireplace Qu', 
                     'Garage Type', 
                     'Garage Finish', 
                     'Garage Qual', 
                     'Garage Cond', 
                     'Fence'], 'NA')

# Check if there is any null values.
null_values = pd.DataFrame(df.isnull().sum())
null_values[null_values[0] != 0]
```

```python
# Rename the column with space or uppercase letters.
# Replace space to "_" and make all column name lowercase.
df.rename(columns=lambda x: x.replace(" ", "_").lower(), inplace = True)
```
