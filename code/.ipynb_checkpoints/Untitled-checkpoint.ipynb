```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, ExtraTreesClassifier
```

```python
df = pd.read_csv('../datasets/train.csv')
```

```python
df.head()
```

```python
df['SalePrice'].median()
```

```python
df['gt_median'] = df.apply(lambda row: 1 if row.SalePrice >= df['SalePrice'].median() else 0, axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df[['Lot Area', 'Overall Qual', 'Bedroom AbvGr']]
y = df['gt_median']
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape
X_test.shape
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
tree.score(X_train, y_train)
```

```python
tree.score(X_test, y_test)
```

### Let's try Bagging!

```python
bagged = BaggingClassifier(base_estimator=DecisionTreeClassifier())
bagged.fit(X_train, y_train)
```

```python
bagged.estimators_[0].tree_.max_depth
```

```python
bagged.score(X_train, y_train)
```

```python
bagged.score(X_test, y_test)
```

### Let's try RandomForest

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(X_train, y_train)
```

```python
temp = rf.estimators_[0]
```

```python
temp.max_features_
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

### Let's try ExtraTrees!

```python
et = ExtraTreesClassifier(max_features=2)
```

```python
et.fit(X_train, y_train)
```

```python
et.score(X_train, y_train)
```

```python
et.score(X_test, y_test)
```

### Let's optimize RandomForest.

```python
from sklearn.model_selection import GridSearchCV
```

```python
my_params = {
    'criterion': ['gini', 'entropy'],    
    'n_estimators': [3, 5, 10, 15, 20],
    'max_features': ['auto', 2, 3],
    'max_depth': [5, 6, 7, 8, 9]
}
```

```python
grid = GridSearchCV(RandomForestClassifier(), my_params, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.score(X_test, y_test)
```

```python
grid.score(X_train, y_train)
```
